# AgriSense — Exploratory Data Analysis

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns; sns.set_theme(style='whitegrid')
```

```python
df = pd.read_csv('../data/raw/Crop_recommendation.csv'); print(df.shape); display(df.head()); display(df.describe())
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(9,7)); sns.heatmap(df.drop('label',axis=1).corr(), annot=True, fmt='.2f', cmap='Greens'); plt.title('Feature Correlations'); plt.tight_layout(); plt.show()
```

```python
df['label'].value_counts().plot(kind='bar', figsize=(14,4), color='#4CAF6A'); plt.title('Crop Class Distribution'); plt.xticks(rotation=45, ha='right'); plt.tight_layout(); plt.show()
```

```python
top5=df['label'].value_counts().head(5).index; sub=df[df['label'].isin(top5)]; [sns.boxplot(data=sub,x='label',y=c,color='#4CAF6A') or plt.title(c) or plt.show() for c in ['N','P','K','temperature','humidity']]
```

```python
y_df = pd.read_csv('../data/raw/yield_df.csv'); print(y_df.shape); display(y_df.head())
```

```python
(y_df['hg/ha_yield']/10000).hist(bins=50, color='#2A7A42'); plt.title('Yield Distribution (t/ha)'); plt.tight_layout(); plt.show()
```

```python
avg=(y_df.groupby('Item')['hg/ha_yield'].mean()/10000).sort_values(); avg.plot(kind='barh', figsize=(9,8), color='#4CAF6A'); plt.title('Avg Yield by Crop (t/ha)'); plt.tight_layout(); plt.show()
```
